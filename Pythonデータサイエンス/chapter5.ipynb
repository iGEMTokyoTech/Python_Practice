```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("ggplot")
plt.rcParams["font.size"] = 14
plt.rcParams["figure.figsize"] = 8, 6
```

```python
dat = pd.read_csv("income_expenses.csv")
x = dat["income"]
y = dat["expenses"]
plt.plot(x, y, "o")
plt.xlim(0, 1200)
plt.ylim(0, 1200)
plt.xlabel("income [jpy]")
plt.ylabel("expenses [jpy]")
plt.show()
```

```python
dat2 = pd.read_csv("proteinanalysis_B2.csv")
x = dat2["Concentration"]
y = dat2["Absorbance"]
plt.plot(x, y, "o")
plt.xlim(0, 100)
plt.ylim(0, 0.25)
plt.show()
```

```python
#B２学生実験：タンパク質定量の結果を用いて、線形回帰のグラフを描画

from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(x[:, np.newaxis], y)

xx = np.array([0,400,50])[:, np.newaxis]
model.predict(xx)

xx= np.arange(0, 200, 100)
yy = model.predict(xx[:, np.newaxis])
plt.plot(xx, yy, label="predicted")
plt.plot(x, y, "o", label="observed")
plt.xlabel("concentration")
plt.ylabel("absorbance")
plt.xlim(0, 100)
plt.ylim(0, 0.25)
plt.legend()
```

```python
corr = np.corrcoef(x, y)
print(corr)
```

```python
# 線形回帰のライブラリ

from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(x[:, np.newaxis], y)
```

```python
print(model.coef_, model.intercept_)
```

```python
print(model.predict([600]))
```

```python
xx = np.array([300,400,500])[:, np.newaxis]
model.predict(xx)
```

```python
xx= np.arange(100, 1200, 100)
yy = model.predict(xx[:, np.newaxis])
plt.plot(xx, yy, label="predicted")
plt.plot(x, y, "o", label="observed")
plt.xlabel("income [jpy]")
plt.ylabel("expenses [jpy]")
plt.xlim(0, 1200)
plt.ylim(0, 1200)
plt.legend()
```

```python
print(model.score(x[:, np.newaxis], y))
```

```python
data = pd.read_csv("blood_fat.csv")
data.head(5)
```

```python
X = data.copy()
y = X.pop("blood fat")
model = linear_model.LinearRegression().fit(X, y)
```

```python
y_pred = model.predict(X)
plt.plot(np.linspace(min(y), max(y), 100), np.linspace(min(y), max(y), 100))
plt.plot(y, y_pred, "o")
plt.xlabel("y")
plt.ylabel("y_pred")
```

```python
print(model.score(X, y))
```
